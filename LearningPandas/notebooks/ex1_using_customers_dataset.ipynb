```python
# Command lets us use the python library pandas with an alias of 'pd'
import pandas as pd

# Create Data Frame (table) from csv file using its relative path
customers_dirty = pd.read_csv("../practice_csv_files/radius_customers_dirty.csv")

# Get the first five rows of the data frame
customers_dirty.head()
```

```python
# Show the shape of the data. In this case 515 rows and 27 columns. Multiplying the two numbers we see we have 13,905 entries or values
customers_dirty.shape
```

```python
# Using the index-based selection operator, iloc, we get the last 5 rows (-5:) for the first 15 columns (0:15)
# In pandas it is row first, column second
customers_dirty.iloc[-5:, 0:15]
```

```python
# Using loc we get the first 15 rows for the columns, 'first_name', 'last_name', and 'age'
customers_dirty.loc[0:15, ['first_name', 'last_name', 'age']]
```

```python
# We can use conditionals to get rows that correspond to what we are looking for. In this case using == and the % (and) operator we get
# rows in which the person's favorite activity is 'Cycling', and their age is over 20
customers_dirty.loc[(customers_dirty.favorite_activities == 'Cycling') & (customers_dirty.age > 20)]
```

```python
# We can use .isin to get the rows that correspond to a list of values we give. In this case those rows (people) that have an age of 22 or 26
# Not a range as it doesn't get values in between the two only 22 or 26
customers_dirty.loc[customers_dirty.age.isin([22.0, 26.0])]
```

```python
# To see rows in which one column is not null we can use .notnull(). Here we see 111 rows in which a value is entered for their favorite activity
# The rest of the rows are ignored as they had null values in this column
customers_dirty.loc[customers_dirty.favorite_activities.notnull()]
```

```python
# Creating a simple data frame to observe we can change a column's existing values by doing df_name['column_name] = 'new value'
reviews = pd.DataFrame({'critic': ['no', 'yes']})
reviews['critic'] = 'everyone'
reviews['critic']
```

```python
# We can get the first 6 rows (loc is inclusive therefore :5 means include row index 5) and get the specified columns for the rows
# Columns can be out of order. In the data frame first_name comes before age and gender, but here we can change the order
customers_dirty.loc[:5, ['age', 'gender', 'first_name']]
```
